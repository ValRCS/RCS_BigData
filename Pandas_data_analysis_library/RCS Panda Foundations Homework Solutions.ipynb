###  Find 25 best(IMDB rating) rated black and white movies that grossed at least 20M
* Sort by Rating first then gross
* Sort by 'imdb_score' first then 'gross'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
film = pd.read_csv('data/movie.csv') #films or movies would be more accurate but we are lazy...
film.head()
```

```python
film.columns
```

```python
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.set_index.html
film=film.set_index('movie_title')
film.head()
```

```python
# https://chrisalbon.com/python/data_wrangling/pandas_list_unique_values_in_column/
film.color.unique()
```

```python
bw = film[film.color == 'Black and White']
```

```python
len(bw)
```

```python
bw.gross[:5]
```

```python
bw[:5]
```

```python
bst=bw[bw.gross > 20000000]
len(bst)
```

```python
bst.head()
```

```python
bst.tail()
```

```python
sortedmovies = bst.sort_values(by=['imdb_score','gross']) # so in case of equal imdb_score gross is the tiebreaker
sortedmovies.head()
```

```python
sortedfilms = bst.sort_values(by=['imdb_score','gross'],ascending=False) # so in case of equal imdb_score gross is the tiebreaker
sortedfilms.head(25)
```

```python
cols = list(sortedfilms.columns)
len(cols),cols
```

```python
# list remove https://docs.python.org/3.1/tutorial/datastructures.html
# better https://stackoverflow.com/questions/36268749/remove-multiple-items-from-a-python-list-in-just-one-statement
# could just copy paste a new list using an existing list if list is short enough..
```

```python
col2 = ['imdb_score','gross','title_year'] + [el for el in cols if el not in ['imdb_score','gross','title_year']]
len(col2),col2
```

```python
bestbwfilms = sortedfilms[col2]
```

```python
bestbwfilms.head(25)
```

```python
bestbwfilms.head(25).to_excel('bestbwfilms.xlsx')
```

```python
df = pd.read_csv('data/movie.csv',index_col='movie_title')
df.head()
```
