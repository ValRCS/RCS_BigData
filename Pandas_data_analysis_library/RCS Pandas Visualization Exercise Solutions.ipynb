```python
import pandas as pd
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
#Create a a new dataframe df2 , 20 rows , 4 columns, 
# filled with random integers from 1 to 6, columns named W,X,Y,Z, index is date starting from today and daily
# \ is used to indicate new line for long expressions
df = pd.DataFrame(np.random.randint(low=1, high=7, size=(20, 4)), columns=list("WXYZ"),\
                  index=pd.date_range('20180830', periods=20, freq="D"))
# using 20 twice is a bit of a code smell a variable/constant would be well suited here
```

```python
len(df)
```

```python
df
```

```python
# create a Bar plot for X and Y columns
df[['X', 'Y']].plot.bar()
```

```python
#save as Bar.pdf
barfig = df[['X', 'Y']].plot.bar().get_figure()
barfig.savefig('Bar.pdf')
```

```python
#create a Box plot for Y,Z columns
#save as Box.png
boxfig = df[['X', 'Y']].plot.box().get_figure()
boxfig.savefig('Box.png')
```

```python
df.columns
```

```python
#create a Line plot for first 3 columns
#save as Line.jpg
linefig = df[df.columns[:3]].plot.line().get_figure()
linefig.savefig('Line.png')
```

```python
# save your data as mydata.xlsx
df.to_excel('mydata.xlsx')
```
